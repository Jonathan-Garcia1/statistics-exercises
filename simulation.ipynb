```python
rng = np.random.default_rng(seed = 42)
```

1 How likely is it that you roll doubles when rolling two dice?

```python
roll = rng.choice([1,2,3,4,5,6],(1_000,2))
roll[:5]
```

```python
double_roll = (roll[:,0] == roll[:,1]).sum()
```

```python
double_roll / len(roll)
```

2 If you flip 8 coins, what is the probability of getting exactly 3 heads? What is the probability of getting more than 3 heads?



```python
flips = rng.choice([0,1],(1_000,8))
flips[:25]
```

```python
triple_h = (flips == 1).sum(axis=1) == 3
triple_h[:5]
```

```python
total_triple = triple_h.sum()
```

```python
total_triple / len(flips)
```

5 Compare Heights

Men have an average height of 178 cm and standard deviation of 8cm.
Women have a mean of 170, sd = 6cm.
Since you have means and standard deviations, you can use np.random.normal to generate observations.
If a man and woman are chosen at random, what is the likelihood the woman is taller than the man?

```python
manh = np.random.normal(178, 8, (1_000,1))
womanh = np.random.normal(170, 6, (1_000,1))
```

```python
manh
```

```python
womanh.shape
```

```python
taller_w = (manh < womanh).sum()
taller_w
```

```python
taller_w / len(manh)
```

```python
heights = np.column_stack((manh, womanh))
heights
```

```python
(heights[:,0] < heights[:,1]).sum()
```

When installing anaconda on a student's computer, there's a 1 in 250 chance that the download is corrupted and the installation fails. What are the odds that after having 50 students download anaconda, no one has an installation issue? 100 students?

```python
corrupt = 1/250
corrupt
```

```python
installs_fifty = rng.random((1_000,50))
installs_fifty[:1]
```

```python
corrumption_fifty = (installs_fifty <= corrupt).sum(axis= 1)
fifty_total = corrumption_fifty.sum()
```

50 Installs

```python
1 - (fifty_total / len(installs_fifty))
```

100 installs

```python
installs_hund = rng.random((1_000,100))
```

```python
corrumption_hund = (installs_hund <= corrupt).sum(axis= 1)
hun_total = corrumption_hund.sum()
1 - (hun_total / len(installs_hund))
```
